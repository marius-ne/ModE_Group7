# Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Visualize demand

```python

# Load the demand data from the CSV in the same folder as this notebook.
df = pd.read_csv("energy_demands.csv")

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df["time [h]"], df["hourly electricity demand [kW]"], label="Electricity demand")
ax.plot(df["time [h]"], df["hourly heat demand [kW]"], label="Heat demand")

max_time = int(df["time [h]"].max())
y_max = max(df["hourly electricity demand [kW]"].max(), df["hourly heat demand [kW]"].max())

for boundary in range(24, max_time + 1, 24):
    ax.axvline(boundary, color="gray", linestyle="--", linewidth=0.8, alpha=0.5)
    ax.text(boundary - 12, y_max * 1.02, f"Day {boundary // 24}", ha="center", va="bottom", fontsize=9, color="gray")

ax.set_xlabel("Time [h]")
ax.set_ylabel("Demand [kW]")
ax.set_title("Energy demands over time")
ax.set_ylim(top=y_max * 1.08)
ax.grid(True, alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```
